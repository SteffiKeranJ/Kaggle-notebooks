```python
import pandas as pd

train = pd.read_csv('train.csv', index_col='Id')
train.head()
```

```python
test = pd.read_csv('test.csv', index_col='Id')
test.head()
```

```python
y = train.SalePrice
y
```

```python
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']

X= train[features].copy()
X
```

```python
X_test=test[features].copy()
X_test
```

```python
from sklearn.model_selection import train_test_split
# Break off validation set from training data
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2,
                                                      random_state=0)
```

```python
from sklearn.ensemble import GradientBoostingRegressor

model_6 = GradientBoostingRegressor(random_state=0)
model_7 = GradientBoostingRegressor(learning_rate=0.15,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsRegressor

model_8 = KNeighborsRegressor()
```

```python
from sklearn.neural_network import MLPRegressor

model_9 = MLPRegressor(random_state=0,batch_size=50, learning_rate='constant', activation='relu',learning_rate_init=0.15)
```

```python
from sklearn.ensemble import RandomForestRegressor

# Define the models
model_1 = RandomForestRegressor(n_estimators=50, random_state=0)
model_2 = RandomForestRegressor(n_estimators=100, random_state=0)
model_3 = RandomForestRegressor(n_estimators=200, criterion='mae', random_state=0)
model_4 = RandomForestRegressor(n_estimators=200, criterion='mae', random_state=9)
model_5 = RandomForestRegressor(n_estimators=200, criterion='mae', random_state=8)

models = [model_3, model_9]
```

```python
from sklearn.metrics import mean_absolute_error

# Function for comparing different models
def score_model(model, X_t=X_train, X_v=X_valid, y_t=y_train, y_v=y_valid):
    model.fit(X_t, y_t)
    preds = model.predict(X_v)
    return mean_absolute_error(y_v, preds)

for i in range(0, len(models)):
    mae = score_model(models[i])
    print("Model %d MAE: %d" % (i+1, mae))
```

```python
# Fit the model to the training data

model_9.fit(X, y)

# Generate test predictions
preds_test = model_9.predict(X_test)

# Save predictions in format used for competition scoring
output = pd.DataFrame({'Id': X_test.index,
                       'SalePrice': preds_test})
output.to_csv('submission.csv', index=False)
```
